

PyTorch also has a bunch of built-in functions like torch.mul() (short for multiplcation) and torch.add() to perform basic operations.


```python
torch.multiply(tensor_add, 10)
```

```python
# Original tensor is still unchanged 
tensor_add
```

```python
# Element-wise multiplication (each element multiplies its equivalent, index 0->0, 1->1, 2->2)
print( tensor_add, "*", tensor_add)
print("Equals:", tensor_add * tensor_add)
```

```python
#Matrix multiplication
torch.matmul(tensor_add, tensor_add), tensor_add @ tensor_add
```

```python
# Shapes need to be in the right way  
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)
```

```python
print(f"Original Shape of tensor_A is {tensor_A} and that of tensor_B is {tensor_B}")
print(f"New shapes for multiplication are \n {tensor_A} and \n {tensor_B.T} where one matrix is transposed so that inner dimensions match\n" )
output = tensor_A @ tensor_B.T
print(output)
print(f"Output shape is {output.shape}")
```

```python
# torch.mm is a shortcut for matmul
torch.mm(tensor_A, tensor_B.T)
```

```python
# Since the linear layer starts with a random weights matrix, let's make it reproducible (more on this later)
torch.manual_seed(42)

# This uses matrix mutliplcation
linear = torch.nn.Linear(in_features=2, out_features=6) # in_features = matches inner dimension of input # out_features = describes outer value 

x= tensor_A
output=linear(x)
print(f"Input shape: {x.shape}\n")
print(f"Output:\n{output}\n\nOutput shape: {output.shape}")
```

Reshaping, stacking, squeezing and unsqueezing

1. torch.reshape(input, shape) -	Reshapes input to shape (if compatible), can also use torch.Tensor.reshape().
2. torch.Tensor.view(shape) - Returns a view of the original tensor in a different shape but shares the same data as the original tensor.
3. torch.stack(tensors, dim=0)- Concatenates a sequence of tensors along a new dimension (dim), all tensors must be same size.
4. torch.squeeze(input) 	- Squeezes input to remove all the dimenions with value 1.
5. torch.unsqueeze(input, dim) - Returns input with a dimension value of 1 added at dim.
6. torch.permute(input, dims) -	Returns a view of the original input with its dimensions permuted (rearranged) to dims.

```python
z = torch.arange(1., 8.)
z, z.shape
```

```python
z_reshaped = z.reshape(1,7)
z_reshaped, z_reshaped.shape
```

```python
# Change view (keeps same data as original but changes view)
# See more: https://stackoverflow.com/a/54507446/7900723
z_view = z.view(1,7)
z_view, z_view.shape
```

```python
# Changing z_view changes z
z_view[:, 0] = 5
z_view, z 
```

```python
# Stack tensors on top of each other
z_stack = torch.stack([z,z,z,z,z], dim=1)
z_stack
```

```python
print(f"Previous tensor: {z_reshaped}")
print(f"Previous shape: {z_reshaped.shape}")
```

```python
# Remove extra dimension from x_reshaped
z_squeezed = z_reshaped.squeeze()
z_squeezed, z_squeezed.shape
```

```python
## Add an extra dimension with unsqueeze
z_unsqueezed = z_squeezed.unsqueeze(dim=0)
z_unsqueezed, z_unsqueezed.shape
```

```python
# permute to a different shape. # Create tensor with specific shape
a = torch.rand(size=(3, 224, 234))
a_permute = a.permute(2, 1, 0)

a.shape, a_permute.shape
```
